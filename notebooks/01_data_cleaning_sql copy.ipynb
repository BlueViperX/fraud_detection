```python
import os    #Used to interact with the operating system.  #Read environment variables #Work with file paths #Access folders/files

import pandas as pd 
import numpy as np
from dotenv import load_dotenv                   # Load environment variables from .env file
from sqlalchemy import create_engine             # Used to create a database connection (MySQL, PostgreSQL, SQL Server, etc.).

load_dotenv()                                    # Load environment variables from .env file

pd.set_option('display.max_columns',100)       # Display all columns in the DataFrame
pd.set_option('display.width',200)

#MY SQL connection
engine = create_engine(os.getenv("DATABASE_URL"))
print("setup complete")          # Display all rows in the DataFrame
```

## LOAD DATA FROM MY SQL

```python
print("loading transaction from sql...")
tx = pd.read_sql("SELECT * FROM transactions", engine)  # Load data from the 'transactions' table in the database
print(f"{tx.shape[0]:,} rows, {tx.shape[1]:,} columns")  # Print the number of rows and columns loaded

print("loading identity from SQL...")
idn = pd.read_sql("SELECT * FROM identity", engine)  # Load data from the 'identity' table in the database
print(f"{idn.shape[0]:,} rows, {idn.shape[1]:,} columns")  # Print the number of rows and columns loaded
```

```python
#Fix empty strings in object columns to NaN

print("Fixing empty strings in object columns to NaN...")
print(f"Before cleaning, transaction data has {tx.isnull().sum().sum():,} missing values and identity data has {idn.isnull().sum().sum():,} missing values.")

for col in tx.select_dtypes(include=['object']).columns:
    tx[col] = tx[col].replace(r'^\s*$', np.nan, regex=True)  # Replace empty strings with NaN in object columns

for col in idn.select_dtypes(include=['object']).columns:
    idn[col] = idn[col].replace(r'^\s*$', np.nan, regex=True)  # Replace empty strings with NaN in object columns    

print(f"After cleaning, transaction data has {tx.isnull().sum().sum():,} missing values and identity data has {idn.isnull().sum().sum():,} missing values.")
```

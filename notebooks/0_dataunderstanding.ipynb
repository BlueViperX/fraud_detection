# Calculating the total Frauds

```python
fraud_count = txn["isFraud"].value_counts()   # Display the count of each class in the 'isFraud' column
fraud_count
```

```python
fraud_percentage = txn["isFraud"].value_counts(normalize=True)*100    # Calculate the percentage of each class in the 'isFraud' column
fraud_percentage   # Calculate the percentage of each class in the 'isFraud' column
```

```python
print(f"Fraud Percentage: {fraud_percentage.iloc[1]}%")     # Print the percentage of fraudulent transactions
```

```python
print("column dtypes:")
print(txn.dtypes.value_counts()) # Display the count of each data type in the DataFrame
```

```python
print("\n Top 10 columns with most missing values:")
print(txn.isnull().sum().sort_values(ascending=False).head(10)) # Dis
```

```python
print("\n Top 10 columns with most missing values:")
missing_values = (txn.isnull().sum()/len(txn)*100).sort_values(ascending=False)
print(missing_values.head(10).round(2))
```
